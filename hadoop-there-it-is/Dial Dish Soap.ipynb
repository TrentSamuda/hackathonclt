```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re

%matplotlib inline
```

```python
df = pd.read_csv('/Users/Adam/Development/hackathonclt/year2/pandas/dial_data.txt', sep='\t', names=['UPC_NUMBER', 'MASTER_UPC_NUMBER', 'ITEM_DESCRIPTION', 'DEPARTMENT_NUMBER', 'DEPARTMENT_DESCRIPTION', 'CATEGORY_NUMBER', 'CATEGORY_DESCRIPTION', 'SUBCATEGORY_NUMBER', 'SUBCATEGORY_DESCRIPTION', 'RECEIPT_NUMBER', 'ITEM_QUANTITY', 'EXTENDED_PRICE_AMOUNT', 'DISCOUNT_QUANTITY', 'EXTENDED_DISCOUNT_AMOUNT', 'TENDER_AMOUNT', 'TRANSACTION_DATETIME', 'EXPRESS_LANE', 'HHID'], header=None)
```

```python
df['trans_datetime'] = pd.to_datetime(df.TRANSACTION_DATETIME)
df['trans_date'] = df['trans_datetime'].map(lambda x: x.strftime('%Y-%m-%d'))
df['trans_weekday'] = df['trans_datetime'].map(lambda x: x.strftime('%A'))
df['trans_hour'] = df['trans_datetime'].map(lambda x: x.strftime('%H'))
```

```python
df.head()
```

```python
df['total_transaction'] = df['ITEM_QUANTITY'] * df['EXTENDED_PRICE_AMOUNT']
df['total_discount'] = df['DISCOUNT_QUANTITY'] * df['EXTENDED_DISCOUNT_AMOUNT']
df['total_sale_amt'] = df['total_transaction'] - df['total_discount']
#df[df['total_sale_amt'] < 0].count()
#df = df[df['total_sale_amt'] > 0]
```

```python
express = df[df['EXPRESS_LANE'] == 1]
non_express = df[df['EXPRESS_LANE'] == 0]
```

```python
summary_data = df.groupby(df['trans_date'])[['total_discount', 'total_sale_amt']].sum()
summary_data.sort(['total_discount', 'total_sale_amt'], ascending=[0, 0])
summary_data.plot(figsize=(20, 10))
```

```python
express_summary = express.groupby(express['trans_date'])['total_sale_amt'].sum()#[['total_sale_amt', 'total_transaction', 'total_discount']].sum()
express_summary.plot(figsize=(20,10))
#express_summary.sort(ascending=0)
#express_summary
```

```python
non_express_summary = non_express.groupby(non_express['trans_date'])['total_sale_amt'].sum()#[['total_sale_amt', 'total_transaction', 'total_discount']].sum()
non_express_summary.plot(figsize=(20,10))
#non_express_summary.sort(ascending=0)
#non_express_summary
```

```python
express_discount = express.groupby(express['trans_date'])['total_discount'].sum()
express_discount.plot(figsize=(20,10))
#express_discount.sort(ascending=0)
#express_discount
```

```python
non_express_discount = non_express.groupby(non_express['trans_date'])['total_discount'].sum()
non_express_discount.plot(figsize=(20,10))
#non_express_discount.sort(ascending=0)
#non_express_discount
```

```python
df_weekday = df.groupby(df['trans_weekday'])[['total_sale_amt', 'total_discount', 'total_transaction']].sum()
df_weekday.plot(figsize=(20,10))
#df_weekday.sort(['total_discount'],ascending=[0])
#df_weekday
```

```python
express_weekday = express.groupby(express['trans_weekday'])[['total_sale_amt', 'total_discount', 'total_transaction']].sum()
express_weekday.plot(figsize=(20,10))
#express_weekday.sort(['total_discount'],ascending=[0])
#express_weekday
```

```python
non_express_weekday = non_express.groupby(non_express['trans_weekday'])[['total_sale_amt', 'total_discount', 'total_transaction']].sum()
non_express_weekday.plot(figsize=(20,10))
#non_express_weekday.sort(['total_discount'],ascending=[0])
#non_express_weekday
```
