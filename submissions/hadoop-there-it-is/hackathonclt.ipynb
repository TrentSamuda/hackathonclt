```python
total_express = dept_express[dept_express['express_lane'] == 1]['freq'].sum()
total_non_express = dept_express[dept_express['express_lane'] == 0]['freq'].sum()
```

```python
express = dept_express[dept_express['express_lane'] == 1]
non_express = dept_express[dept_express['express_lane'] == 0]
express['pct'] = express['freq']/total_express
non_express['pct'] = non_express['freq']/total_non_express

#pd.merge(express, non_express)
```

```python
dept_express = pd.concat([express, non_express])
dept_express[dept_express['dept'] == 'DAIRY']
```
